```python
from collections import deque
import math
```

```python
# Declaring lists
numbersListSimple = [20,10,30,40,9,49,25,15,33,42]
numbersListComplicated = [1000,5,1,9,100]
```

```python
# InsertionSort alghortim from https://www.geeksforgeeks.org/python-program-for-insertion-sort/
def insertionSort(arr):
    n = len(arr)  #Get the length of the array
      
    if n <= 1:
        return  #If the array has 0 or 1 element, it is already sorted, so return
 
    for i in range(1, n):  #Iterate over the array starting from the second element
        key = arr[i]  #Store the current element as the key to be inserted in the right position
        j = i-1
        while j >= 0 and key < arr[j]:  #Move elements greater than key one position ahead
            arr[j+1] = arr[j]  #Shift elements to the right
            j -= 1
        arr[j+1] = key  #Insert the key in the correct position
```

```python
# BucketSort alghoritm
def BucketSort(list):
    M = max(list)
    K = len(list)
    k_lists = [deque() for index in range(K)]
    for number in list:
        k_lists[math.floor((K*number)/(M+1))].append(number)
    for k_list in k_lists:
        insertionSort(k_list)
    result_list = []
    for sortedNumber in k_lists:
        result_list.extend(sortedNumber)
    print(result_list)
```

```python
BucketSort(numbersListComplicated)
BucketSort(numbersListSimple)
```

# Test cases
#### Generated using random function in python. Saving run times and list sizes in a seperate lists

```python
import random
import time
import matplotlib.pyplot as plt
```

```python
# Scenario 1: Input numbers are evenly distributed
# Simple lists
simpleList_times = []
simpleList_sizes = []

for list in range(200):
    list_size = random.randint(100, 1000)
    randomList = random.choices(range(1, list_size), k=list_size)  # Evenly distributed numbers thanks to random.choices() method
                                                                # https://www.w3schools.com/python/ref_random_choice.asp
    start_time = time.time()
    BucketSort(randomList)
    simpleList_times.append(time.time() - start_time)
    simpleList_sizes.append(list_size)
```

```python
# Scenario 2: Input numbers are not evenly distributed
#Complicated lists
complicatedList_times = []
complicatedList_sizes = []

for list in range(200):
    list_size = random.randint(100, 1000)
    randomList = random.choices(range(1, list_size), k=list_size)
    #Making the list uneven thanks to adding huge number
    randomList.pop()
    randomList.append(1000000)
    start_time = time.time()
    BucketSort(randomList)
    complicatedList_times.append(time.time() - start_time)
    complicatedList_sizes.append(list_size)
```

# Graph depicting the runtime of BucketSort as a function of input size for both scenarios

```python
plt.figure(figsize=(10, 8))
plt.scatter(simpleList_sizes, simpleList_times, label='SimpleList (bucketSort)', marker='o', color='blue')
plt.scatter(complicatedList_sizes, complicatedList_times, label='ComplicatedList (bucketSort)', marker='o', color='orange')
plt.xlabel('Input size')
plt.ylabel('Time')
plt.legend()
plt.show()
```

# Conclusion based on graph
The BucketSort alghoritm has O(n) Time complexity for simpleList, which contains list of K, evenly distributed numbers. Whereas, for complicatedList, which contains list of K elements unevenly distirbuted, the time complexity is O(n^2). In complicatedList, the execution time rises consequently when the input size expands. 
### Summing up
The BucketSort alghoritm is most efficient in case of lists with evenly distirbuted numbers.
